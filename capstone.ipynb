# Function to retrieve NBA Bios


```python
def get_nba_bios(url):
    response = requests.get(url)
    soup = BS(response.text)

    #nba bio headers
    nba_bio_headers = soup.findAll('th', class_='poptip')
    nba_bio_headers = [name.text for name in nba_bio_headers]
    nba_bio_headers

    #nba player names
    nba_bio_name = soup.findAll('th', class_='left')
    nba_bio_name = [name.text for name in nba_bio_name]
    nba_bio_name


    #nba debut year 
    nba_debut = soup.findAll('td', attrs={'data-stat':'year_min'})
    nba_debut = [debut.text for debut in nba_debut]
    nba_debut

    #nba final year 
    nba_final_year = soup.findAll('td', attrs={'data-stat':'year_max'})
    nba_final_year = [debut.text for debut in nba_final_year]
    nba_final_year


    #nba position
    nba_pos = soup.findAll('td', attrs={'data-stat':'pos'})
    nba_pos = [final.text for final in nba_pos]
    nba_pos

    #nba height
    nba_height = soup.findAll('td', attrs={'data-stat':'height'})
    nba_height = [height.text for height in nba_height]
    nba_height

    #nba weight
    nba_weight = soup.findAll('td', attrs={'data-stat':'weight'})
    nba_weight = [weight.text for weight in nba_weight]
    nba_weight
    
    #birthdate
    birthdate = soup.findAll('td', attrs={'data-stat':'birth_date'})
    birthdate = [date.text for date in birthdate]
    birthdate
    
    #college
    college = soup.findAll('td', attrs={'data-stat':'colleges'})
    college = [school.text for school in college]
    college
    
    nba_bio = pd.DataFrame({'Player': nba_bio_name, 'From': nba_debut, 'To':nba_final_year, 'Pos':nba_pos, 'Ht':nba_height, 'Wt':nba_weight, 'Birth Date': birthdate, 'Colleges':college})

    return nba_bio
```

# Function to Find Player's Site URL

```python
def player_site(first, last):
    first_name = first
    last_name = last
    url = f'https://www.basketball-reference.com/search/search.fcgi?hint={first_name}+{last_name}&search={first_name}+{last_name}&pid=&idx='
    response = requests.get(url)
    soup = BS(response.text)
    url = 'https://www.basketball-reference.com' + soup.find('div', class_ = 'search-item-url').text
    return url
```

# Function to find NBA stats on Individual Player Sites

```python
def find_player_site(first, last):
    letter =last[0].lower()
    player_name = first + ' ' + last
    url = f'https://www.basketball-reference.com/players/{letter}'
    response = requests.get(url)
    soup = BS(response.content, "html.parser")
    player_url = soup.findAll('a', string = player_name)
    player_url = player_url[0]['href']
    player_url = 'https://www.basketball-reference.com' + player_url
    response = requests.get(player_url)
    # Parse HTML content
    soup = BS(response.content, "html.parser")
    # Find the table with the specified ID
    table = soup.find("table", {"id": "per_game"})
    tfoot = table.find("tfoot")
    career_row = [td.get_text() for td in tfoot.find_all("td")]
    return career_row[4:29]
```

```python
draft_columns = ['Pk', 'Tm', 'Player', 'College', 'Yrs', 'G', 'MP', 'PTS', 'TRB', 'AST', 'FG%', '3P%', 'FT%',
                 'MP', 'PTS', 'TRB', 'AST', 'WS', 'WS/48', 'BPM', 'VORP', 'Draft Year']
draft = pd.DataFrame(columns=draft_columns)
```

## Years of NBA Draft 

```python
years = ['2013','2014','2015','2016','2017','2018','2019','2020', '2021','2022','2023']
```

## Function to find the draft position of each player

```python
def draft_info(draft_year):
    url = f'https://www.basketball-reference.com/draft/NBA_{draft_year}.html'
    response = requests.get(url)
    soup = BS(response.text, 'html.parser')
    column_data = soup.findAll('tr')

    draft_year_data = []

    for row in column_data:
        row_data = row.findAll('td')
        individual_row_data = [data.text.strip() for data in row_data]
        if len(individual_row_data) > 1:
            individual_row_data.append(draft_year)
            draft_year_data.append(individual_row_data)

    return pd.DataFrame(draft_year_data, columns=draft_columns)


for year in years:
    draft_year_df = draft_info(year)
    draft = pd.concat([draft, draft_year_df], ignore_index=True)
    time.sleep(4)

    #saved draft to csv named "draft"
```
